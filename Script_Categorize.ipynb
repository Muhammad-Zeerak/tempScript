```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
temp = [] # list to keep the keywords that are useful

from rake_nltk import Rake
rake_nltk_var = Rake()
for text in df['Meta Tag']:
  rake_nltk_var.extract_keywords_from_text(str(text))
  keyword_extracted = rake_nltk_var.get_ranked_phrases()
  if(len(keyword_extracted) > 0):
    temp.append(keyword_extracted[0])
```

```python
len(temp)
temp
```

```python
import re
 
# using regular expression to remove special characters
out_list = [re.sub(r'[^a-zA-Z0-9 ]','',string) for string in temp]

# print list without special characters
len(out_list)
```
